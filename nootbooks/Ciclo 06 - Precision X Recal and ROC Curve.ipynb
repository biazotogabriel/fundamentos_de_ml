# Imports

```python
from sklearn import datasets as dt
import matplotlib.pyplot as plt
from sklearn import model_selection as ms
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics as mt
```

# Load Dataset

```python
X, y = dt.make_classification(n_samples=500,
                              n_classes=2,
                              n_redundant=5,
                              random_state=0)
```

```python
#plt.scatter(X[:, 0], X[:, 1], c=y);
```

```python
X_train, X_test, y_train, y_test = ms.train_test_split(X, y, test_size=0.2)
```

```python
# define
dtc_model = DecisionTreeClassifier(max_depth=4)

# fit
dtc_model.fit(X_train, y_train)

# predict
y_pred = dtc_model.predict_proba(X_test)
y_pred = y_pred[:, 1]
precision, recall, thresholds = mt.precision_recall_curve(y_test, y_pred)
print(precision)
```

# Precision X Recall Curve

```python
plt.plot(recall, precision, 'o-', label='Model')
plt.xlabel('recall') 
plt.ylabel('precision')
plt.legend();
```

```python
plt.plot(thresholds, precision[:-1], 'b--', label='Precision')
plt.plot(thresholds, recall[:-1], 'g-', label='Recall')
plt.xlabel('thresholds')
plt.xlabel('precision/recall')
plt.legend()

#plt.vlines(0.45, 0, 1, color='r', linestyles='--')
#plt.hlines(0.9, 0, 1, color='b', linestyles='--')
#plt.hlines(0.86, 0, 1, color='g', linestyles='--')
```

# ROC Curve


```python
fpr, tpr, thresholds = mt.roc_curve(y_test, y_pred)
```

```python
plt.plot(fpr, tpr, 'bo--', label='Thresholds')
#plt.plot(thresholds, recall, 'g-', label='Recall')
plt.xlabel('fpr')
plt.ylabel('tpr')
plt.legend()
for i in range(len(thresholds)):
    plt.text(fpr[i]+0.01, tpr[i]-0.03, f'{thresholds[i]:0.2f}')
```

```python
plt.plot(thresholds, tpr, 'b--', label='tpr')
plt.plot(thresholds, 1-fpr, 'g-', label='fpr')
plt.xlabel('thresholds')
plt.ylabel('tpr/fpr')
plt.legend()
plt.vlines(0.44, 0, 1)
plt.xlim([0, 1])
plt.ylim([0, 1])
```
